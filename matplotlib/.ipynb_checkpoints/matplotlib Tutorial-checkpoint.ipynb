# Matplotlib Tutorial

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

## Plot

```python
x_1 = np.linspace(0, 5,10)
y_1 = x_1**2
plt.plot(x_1,y_1)
plt.title('Days squared chart')
plt.ylabel('days Squared')
plt.xlabel('Days')
```


## Multiple plots

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```
